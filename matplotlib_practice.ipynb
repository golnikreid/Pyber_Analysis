```python
import matplotlib.pyplot as plt
import statistics
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
months = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
total_fare = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

xlabel = "Date"
ylabel = "Fare($)"
title = "Pyber Fare by Month"
```

```python
plt.plot(months, total_fare, label="Boston", marker="*", color="blue", linewidth=2)
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
stdev = statistics.stdev(total_fare)
plt.errorbar(months, total_fare, yerr=stdev, capsize=3)
```

```python
# create a plot with ax.plt()
fig, ax = plt.subplots()

ax.set_title("Pyber Fare by Month")
ax.set_ylabel("Fare($)")
ax.set_xlabel("Date")
ax.legend()
ax.plot(months, total_fare, color="green", marker="D", linewidth=2, label="Boston")
ax.grid()
#ax.show()
```

```python
plt.bar(months, total_fare, color="green", label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("Pyber Fare by Month")
plt.legend()
plt.show()
```

```python
plt.barh(months, total_fare, color="magenta")
plt.legend()
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("Pyber Fare by Month")
plt.gca().invert_yaxis()
```
